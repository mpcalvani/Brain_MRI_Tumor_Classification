```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Percorso base: dalla cartella notebooks risali di uno e vai in data/brain_tumor_mri
base_dir = "../data/brain_tumor_mri"

train_dir = os.path.join(base_dir, "Training")
test_dir = os.path.join(base_dir, "Testing")

# Le 4 classi del dataset
classes = ["glioma", "meningioma", "notumor", "pituitary"]

print("TRAINING SET:")
for cls in classes:
    folder = os.path.join(train_dir, cls)
    n_images = len(os.listdir(folder))
    print(f"  {cls}: {n_images} immagini")

print("\nTESTING SET:")
for cls in classes:
    folder = os.path.join(test_dir, cls)
    n_images = len(os.listdir(folder))
    print(f"  {cls}: {n_images} immagini")

# Mostra un esempio per ogni classe dal training set
plt.figure(figsize=(10, 8))

for i, cls in enumerate(classes):
    folder = os.path.join(train_dir, cls)
    first_img_name = os.listdir(folder)[0]  # prende la prima immagine della cartella
    img_path = os.path.join(folder, first_img_name)

    img = Image.open(img_path)

    plt.subplot(2, 2, i+1)
    plt.imshow(img, cmap="gray")
    plt.title(cls)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
